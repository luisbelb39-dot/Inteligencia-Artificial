## **Tarea 2 McCullock-Pitts**

**Notebook 1: Operaci√≥n OR**

```python
# McCulloch-Pitts para OR
import numpy as np

# Entradas posibles
entradas = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Pesos y umbral
pesos = np.array([1,1])
umbral = 1

# Funci√≥n de activaci√≥n
def mc_or(x):
    return 1 if np.dot(x, pesos) >= umbral else 0

# Resultados
for x in entradas:
    print(f"Entrada: {x}, Salida: {mc_or(x)}")
```

**Explicaci√≥n de McCulloch-Pitts para OR**
"La neurona McCulloch-Pitts recibe dos entradas binarias. Si la suma ponderada de las entradas alcanza el umbral (‚â•1), la salida es 1. Esto representa la operaci√≥n l√≥gica OR."

**Notebook 2: Operaci√≥n NOT**

```python
# McCulloch-Pitts para NOT
import numpy as np

# Entradas posibles
entradas = np.array([0,1])

# Peso y umbral
peso = -1
umbral = 0

def mc_not(x):
    return 1 if (x * peso) >= umbral else 0

# Resultados
for x in entradas:
    print(f"Entrada: {x}, Salida: {mc_not(x)}")
```

**Exolicaci√≥n de Operaci√≥n NOT**
"La neurona McCulloch-Pitts recibe una entrada binaria. Con peso negativo y umbral en 0, cuando la entrada es 1 la salida es 0, y cuando la entrada es 0 la salida es 1. Esto simula la operaci√≥n l√≥gica NOT."

## **Tarea 3 Perpectron**

```python
# Tarea 3: Perceptr√≥n
## Operaci√≥n OR
# Perceptr√≥n para la operaci√≥n OR

import numpy as np

# Datos de entrada (x1, x2)
entradas = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Salidas esperadas (OR)
salidas_esperadas = np.array([0,1,1,1])

# Inicializaci√≥n de pesos y tasa de aprendizaje
pesos = np.random.rand(2)
sesgo = np.random.rand()
tasa_aprendizaje = 0.1

# Funci√≥n de activaci√≥n (escal√≥n)
def activacion(x):
    return 1 if x >= 0 else 0

# Entrenamiento
for epoca in range(10):
    for i in range(len(entradas)):
        entrada = entradas[i]
        salida_real = salidas_esperadas[i]
        suma = np.dot(entrada, pesos) + sesgo
        salida_predicha = activacion(suma)
        error = salida_real - salida_predicha

        # Ajuste de pesos y sesgo
        pesos += tasa_aprendizaje * error * entrada
        sesgo += tasa_aprendizaje * error

print("Pesos finales:", pesos)
print("Sesgo final:", sesgo)

# Pruebas
for entrada in entradas:
    salida = activacion(np.dot(entrada, pesos) + sesgo)
    print(f"Entrada: {entrada}, Salida: {salida}")
```

##**# Explicaci√≥n**
El perceptr√≥n recibe dos entradas binarias y ajusta sus pesos mediante aprendizaje supervisado.  
Durante el entrenamiento, compara su salida con la deseada (tabla OR) y corrige sus pesos hasta que la salida sea correcta.  
Finalmente, el perceptr√≥n logra clasificar correctamente la operaci√≥n OR.

```python
# Perceptr√≥n para la operaci√≥n NOT

import numpy as np

# Entradas (una sola neurona)
entradas = np.array([[0], [1]])

# Salidas esperadas (NOT)
salidas_esperadas = np.array([1, 0])

# Inicializaci√≥n de peso, sesgo y tasa de aprendizaje
peso = np.random.rand()
sesgo = np.random.rand()
tasa_aprendizaje = 0.1

# Funci√≥n de activaci√≥n
def activacion(x):
    return 1 if x >= 0 else 0

# Entrenamiento
for epoca in range(10):
    for i in range(len(entradas)):
        entrada = entradas[i]
        salida_real = salidas_esperadas[i]
        suma = np.dot(entrada, peso) + sesgo
        salida_predicha = activacion(suma)
        error = salida_real - salida_predicha

        # Ajuste de peso y sesgo
        peso += tasa_aprendizaje * error * entrada
        sesgo += tasa_aprendizaje * error

print("Peso final:", peso)
print("Sesgo final:", sesgo)

# Pruebas
for entrada in entradas:
    salida = activacion(np.dot(entrada, peso) + sesgo)
    print(f"Entrada: {entrada}, Salida: {salida}")
```

### Explicaci√≥n
El perceptr√≥n aprende la operaci√≥n NOT ajustando su peso y sesgo.  
Al entrenarlo, logra que cuando la entrada sea 0 la salida sea 1, y cuando la entrada sea 1 la salida sea 0, reproduciendo correctamente la tabla l√≥gica NOT.

```python
import numpy as np
import pandas as pd

# Generar dataset sint√©tico con diferentes rangos por habilidad
n_muestras = 5000
# Habilidades t√©cnicas: 70-100
tec = np.random.randint(70, 101, (n_muestras, 6))
# Habilidades matem√°ticas: 60-90
mat = np.random.randint(60, 91, (n_muestras, 6))
# Habilidades psicol√≥gicas: 50-85
psi = np.random.randint(50, 86, (n_muestras, 6))
# Nombres de columnas
columnas = [
    'python', 'sql', 'visualizacion_datos', 'excel', 'manejo_apis', 'nube',
    'estadistica', 'probabilidad', 'algebra_lineal', 'calculo', 'optimizacion', 'modelado',
    'trabajo_equipo', 'comunicacion', 'creatividad', 'adaptabilidad', 'liderazgo', 'resolucion_problemas'
]
# Crear DataFrame
df = pd.DataFrame(np.concatenate([tec, mat, psi], axis=1), columns=columnas)
```

```python
# Calcular promedio por grupo y promedio total
prom_tecnologicas = df[['python', 'sql', 'visualizacion_datos', 'excel', 'manejo_apis', 'nube']].mean(axis=1)
prom_matematicas  = df[['estadistica', 'probabilidad', 'algebra_lineal', 'calculo', 'optimizacion', 'modelado']].mean(axis=1)
prom_psicologicas = df[['trabajo_equipo', 'comunicacion', 'creatividad', 'adaptabilidad', 'liderazgo', 'resolucion_problemas']].mean(axis=1)
# Promedio total = promedio de los 3 promedios
prom_total = (prom_tecnologicas + prom_matematicas + prom_psicologicas) / 3
# Etiqueta: 1 = apto si promedio total > 80
y = (prom_total > 75).astype(int)
# one-hot encoding
y = pd.get_dummies(y).values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df, y, test_size=0.2, random_state=42
)

# Escalamiento
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
modelo = Sequential([
    Dense(64, input_shape=(X_train.shape[1],), activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(2, activation='softmax')
])

adam = Adam(learning_rate=0.001)
modelo.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#entrenamiento
history = modelo.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    validation_data=(X_test, y_test),
    verbose=1
)
```

```python
#evaluacion
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='P√©rdida de entrenamiento')
plt.plot(history.history['val_loss'], label='P√©rdida de validaci√≥n')
plt.xlabel('√âpocas')
plt.ylabel('P√©rdida')
plt.legend()
plt.title('Evoluci√≥n de la funci√≥n de p√©rdida')
plt.show()

loss, acc = modelo.evaluate(X_test, y_test)
print(f"\nüîπ Loss: {loss:.4f}  |  Accuracy: {acc:.4f}")
```

```python
#matriz de confusion
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, recall_score

pred = modelo.predict(X_test)
y_pred = np.argmax(pred, axis=1)
y_true = np.argmax(y_test, axis=1)

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Apto', 'Apto'])
disp.plot(cmap='Blues')
plt.show()

sensitivity = recall_score(y_true, y_pred, average=None)
print("\nSensitivity (Recall) por clase:")
print(f"No Apto: {sensitivity[0]:.2f}")
print(f"Apto: {sensitivity[1]:.2f}")
```

```python
nuevo_candidato = np.array([[90, 88, 85, 87, 84, 89,   # t√©cnicas
                             82, 86, 85, 83, 87, 80,   # matem√°ticas
                             70, 75, 80, 72, 78, 74]]) # psicol√≥gicas

nuevo_candidato_scaled = scaler.transform(nuevo_candidato)
pred_nuevo = modelo.predict(nuevo_candidato_scaled)
clase_predicha = np.argmax(pred_nuevo)

print("Evaluaci√≥n de nuevo candidato:")
print(f"Resultado: {'Apto' if clase_predicha == 1 else 'No Apto'}")
```
